```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
import pandas as pd
from sklearn.utils import shuffle
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/DT-Wage.csv')
df.head()
```

```python
df.dropna(how="all",axis=1,inplace=True)
df
```

```python
df.dropna(how="all",axis=0,inplace=True)
df
```

```python
df=df.drop_duplicates()
df
```

```python
df= df.replace(r'1.', '', regex=True)
df= df.replace(r'2.', '', regex=True)
df= df.replace(r'3.', '', regex=True)
df= df.replace(r'4.', '', regex=True)
df= df.replace(r'5.', '', regex=True)
df= df.replace(r'<=', '', regex=True)
df= df.replace(r'>=', '', regex=True)
df= df.replace(r'<', '', regex=True)
df= df.replace(r'<', '', regex=True)
```

```python
df
```

```python
df.columns
```

```python
df = df.drop(columns=['region'])
df = df.drop(columns=['year'])
```

```python
df = df.drop(columns=['wage'])
```

```python
df.columns
```

```python
# Shuffle the dataset for randomness
df = shuffle(df, random_state=42).reset_index(drop=True)
```

```python
le=LabelEncoder()
```

```python
# Separate features (X) and labels (y)
X = df.drop(columns="logwage")
y = df["logwage"]
```

```python
# Calculate split indices
train_end = int(0.7 * len(X))
val_end = int(0.85 * len(X))

# Split the data into train, validation, and test sets
X_train, y_train = X[:train_end], y[:train_end]
X_val, y_val = X[train_end:val_end], y[train_end:val_end]
X_test, y_test = X[val_end:], y[val_end:]
```

```python
# Display shapes to verify
print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_val:", X_val.shape, "y_val:", y_val.shape)
print("X_test:", X_test.shape, "y_test:", y_test.shape)
```

```python
y_val
```

```python
# Encode categorical features and labels
label_encoders = {col: LabelEncoder() for col in X_train.columns}

# Transform X
for col in X_train.columns:
    X_train[col] = label_encoders[col].fit_transform(X_train[col])
    X_test[col] = label_encoders[col].fit_transform(X_test[col])
```
