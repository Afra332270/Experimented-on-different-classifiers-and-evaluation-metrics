```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
import pandas as pd
from sklearn.utils import shuffle
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/DT-Credit.csv')
df.head()
```

```python
df.dropna(how="all",axis=1,inplace=True)
df
```

```python
df.dropna(how="all",axis=0,inplace=True)
df
```

```python
df=df.drop_duplicates()
df
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler

columns_to_scale = ['Income', 'Limit', 'Balance']

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform only the selected columns
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
df
```

```python
y = df['Balance']
X = df.drop(columns=['Balance'])
```

```python
le=LabelEncoder()
```

```python
# Applying LabelEncoder on categorical columns
X['Own'] = le.fit_transform(X['Own'])
X['Student'] = le.fit_transform(X['Student'])
X['Married'] = le.fit_transform(X['Married'])
X['Region'] = le.fit_transform(X['Region'])
```

```python
X
```

```python
# Calculate split indices
train_end = int(0.7 * len(X))
val_end = int(0.85 * len(X))

# Split the data into train, validation, and test sets
X_train, y_train = X[:train_end], y[:train_end]
X_val, y_val = X[train_end:val_end], y[train_end:val_end]
X_test, y_test = X[val_end:], y[val_end:]
```

```python
# Display shapes to verify
print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_val:", X_val.shape, "y_val:", y_val.shape)
print("X_test:", X_test.shape, "y_test:", y_test.shape)
```
